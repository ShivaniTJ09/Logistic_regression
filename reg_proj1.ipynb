```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

```python
##data preparation
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS' , axis=1)
x
```

# data Splitting


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x ,y ,test_size=0.2 , random_state= 100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

# Evaluate mode

```python
from numpy.core.fromnumeric import mean
from sklearn.metrics import mean_squared_error, r2_score

lr_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_r2_score = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2_csore = r2_score(y_test, y_lr_test_pred)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_mse, lr_test_mse, lr_r2_score, lr_test_r2_csore]).transpose()
lr_results.columns = ['method','training MSE','trainig R2', 'testing MSE', 'teting_r2']
```

```python
lr_results
```

# Random_forest

```python
from sklearn.ensemble import RandomForestRegressor

rf= RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
rf_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_r2_score = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2_csore = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_mse, rf_test_mse, rf_r2_score, rf_test_r2_csore]).transpose()
rf_results.columns = ['method','training MSE','trainig R2', 'testing MSE', 'teting_r2']
```

```python
rf_results
```

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```
